```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
immi_merge_2 = pd.read_csv('immi_2.csv',index_col=0)
boundary = r'E:\casa\casa0003\new test group work\data\world-administrative-boundaries\world-administrative-boundaries.shp'
boundary_shp = gpd.read_file(boundary)
boundary_shp['centroid'] = boundary_shp['geometry'].centroid
boundary_shp['origin'] = boundary_shp.loc[boundary_shp['name'] == 'China','centroid']
poly_merged = immi_merge_2.merge(boundary_shp, how='inner')
poly_merged = poly_merged.iloc[:,:-2]
poly_merged.head()
```

```python
china_coord = boundary_shp.loc[boundary_shp['name'] == 'China','centroid']
china_x = china_coord.x
china_y = china_coord.y
```

```python
boundary_shp.head()
```

```python
cen_shp = boundary_shp.drop(columns=['geometry','origin'])
point_merged = immi_merge_2.merge(cen_shp, how='inner')
```

```python
point_merged_origin = point_merged.rename(columns={'geometry':'centroid'})
point_merged_origin.head()
```

```python
point_merged = point_merged[['Phase','Time_est','country','Population_est','centroid']]
point_merged.head()
```

```python
from scipy.interpolate import interp1d
import numpy as np
trails_df = pd.DataFrame(columns=['Phase','country','Population_est','trails'])
for i in range(len(point_merged)):
    for j in range(500):
        trails_df = trails_df.append(point_merged.iloc[i,:4],ignore_index=True)
        
    time_s = [point_merged.loc[i,'Time_est']-5, point_merged.loc[i,'Time_est']]
    xs = [china_x, point_merged.loc[i,'centroid'].x]
    ys = [china_y, point_merged.loc[i,'centroid'].y]
    fe_x = interp1d(time_s,xs)
    fe_y = interp1d(time_s,ys)

    trail_x = fe_x(np.arange(point_merged.loc[i,'Time_est'] - 5, point_merged.loc[i,'Time_est'],0.01))
    trail_y = fe_y(np.arange(point_merged.loc[i,'Time_est'] - 5, point_merged.loc[i,'Time_est'],0.01))
    coords = pd.DataFrame(columns=['time','x','y'])
    coords['time'] = pd.DataFrame(np.arange(point_merged.loc[i,'Time_est'] - 5, point_merged.loc[i,'Time_est'],0.01))
    coords['x'] = trail_x
    coords['y'] = trail_y
    point = gpd.GeoDataFrame(coords, geometry=gpd.points_from_xy(coords.x, coords.y)) 
    trails_df.iloc[500*i:500*i+500,-1] = point.iloc[:,-1]
    trails_df.iloc[500*i:500*i+500,-2] = point.iloc[:,0]

trails_df
```

```python
trails_df.to_csv('trails_df.csv')
```
